```python
# --- Imports ---
import os
import zipfile
import shutil
import pandas as pd
import numpy as np
import tensorflow as tf
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Conv2D, MaxPooling2D, Flatten, Dense, Dropout, BatchNormalization # Added BatchNormalization
from tensorflow.keras.preprocessing.image import ImageDataGenerator # Used for test.zip
from tensorflow.keras.utils import to_categorical
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.model_selection import train_test_split
from sklearn.metrics import classification_report, confusion_matrix
from sklearn.preprocessing import LabelBinarizer # For one-hot encoding labels from CSV
from keras.callbacks import ReduceLROnPlateau # From provided Kaggle script
from PIL import Image, ImageDraw, ImageFont # For creating dummy images and text


# --- Configuration ---
train_csv_name = 'train.csv' # Name of your training CSV file
test_zip_name = 'test.zip'   # Name of your testing ZIP file (containing images in folders)
extracted_train_dir_temp = 'extracted_train_data_temp' # Temporary dir if train.csv is inside a zip
extracted_test_dir = 'extracted_test_data'   # Temp dir for test images from test.zip
model_save_path = 'trained_hand_gesture_model.h5' # Path to save the final trained model

# These dimensions are typical for Sign Language MNIST and will be used for both CSV and zipped images
IMG_WIDTH = 28
IMG_HEIGHT = 28
CHANNELS = 1 # 1 for grayscale

BATCH_SIZE = 128 # Used 128 in Kaggle script's model.fit
EPOCHS = 20 # Used 20 in Kaggle script

# Split ratio for train.csv data into training and validation sets
# Test data comes entirely from test.zip
VAL_SPLIT_FROM_TRAIN_CSV = 0.15 # 15% of train.csv data for validation


print("--- Configuration Loaded ---")
print(f"Training data source: '{train_csv_name}'")
print(f"Testing data source: '{test_zip_name}' (expected to contain images in class subfolders)")
print(f"Image dimensions: {IMG_WIDTH}x{IMG_HEIGHT} (Grayscale)")
print(f"Training for {EPOCHS} epochs with batch size {BATCH_SIZE}.")

# --- 1. Dataset Preparation: Create Dummies (if needed), Unzip, and Load Data ---

# Clean up previous extractions
print(f"\n--- Cleaning up previous data directories ---")
if os.path.exists(extracted_train_dir_temp):
    shutil.rmtree(extracted_train_dir_temp)
    print(f"Removed existing temporary '{extracted_train_dir_temp}'.")
if os.path.exists(extracted_test_dir):
    shutil.rmtree(extracted_test_dir)
    print(f"Removed existing temporary '{extracted_test_dir}'.")

# --- Dummy Data Creation ---
# Create dummy train.csv if it doesn't exist
if not os.path.exists(train_csv_name):
    print(f"\n--- Creating a dummy '{train_csv_name}' (CSV based) for demonstration ---")
    print("This dummy mimics Sign Language MNIST CSV structure (labels 0-24, skipping 9/J).")
    num_samples_train_dummy = 500 # More samples for better dummy training

    # Generate ASL-like labels (0-24, skipping 9 for 'J')
    dummy_labels = [i for i in range(25) if i != 9] # A-Z excluding J
    num_dummy_classes = len(dummy_labels) # Should be 24

    pixel_cols = [f'pixel{i}' for i in range(IMG_WIDTH * IMG_HEIGHT)]
    data_train_dummy_list = []
    labels_train_dummy_list = []

    for i in range(num_samples_train_dummy):
        actual_label_idx = i % num_dummy_classes # Cycle through dummy classes
        label_val = dummy_labels[actual_label_idx] # Get the numerical label (e.g., 0, 1, 2, ..., 8, 10, ...)

        row_pixels = np.random.randint(0, 256, size=(IMG_WIDTH * IMG_HEIGHT)).astype(np.uint8)

        # Simple visual pattern for dummy images
        img_2d = row_pixels.reshape(IMG_WIDTH, IMG_HEIGHT)
        center_x, center_y = IMG_WIDTH // 2, IMG_HEIGHT // 2

        # Draw a simple shape based on class_id for visual distinctness
        if label_val % 3 == 0: # Circle
            for x in range(IMG_WIDTH):
                for y in range(IMG_HEIGHT):
                    if (x - center_x)**2 + (y - center_y)**2 < (IMG_WIDTH/3)**2:
                        img_2d[y, x] = np.random.randint(150, 256) # Brighter center
                    else:
                        img_2d[y, x] = np.random.randint(0, 50) # Darker background
        elif label_val % 3 == 1: # Square
            for x in range(IMG_WIDTH):
                for y in range(IMG_HEIGHT):
                    if (IMG_WIDTH*0.2 < x < IMG_WIDTH*0.8) and (IMG_HEIGHT*0.2 < y < IMG_HEIGHT*0.8):
                        img_2d[y, x] = np.random.randint(150, 256)
                    else:
                        img_2d[y, x] = np.random.randint(0, 50)
        else: # Diagonal line
            for x in range(IMG_WIDTH):
                for y in range(IMG_HEIGHT):
                    if abs(x - y) < 5: # Thick diagonal
                        img_2d[y, x] = np.random.randint(150, 256)
                    else:
                        img_2d[y, x] = np.random.randint(0, 50)

        data_train_dummy_list.append(img_2d.flatten())
        labels_train_dummy_list.append(label_val)

    df_train_dummy = pd.DataFrame(data_train_dummy_list, columns=pixel_cols)
    df_train_dummy['label'] = labels_train_dummy_list
    df_train_dummy.to_csv(train_csv_name, index=False)
    print(f"Dummy '{train_csv_name}' created with {num_samples_train_dummy} samples.")

# Create dummy test.zip if it doesn't exist
if not os.path.exists(test_zip_name):
    print(f"\n--- Creating a dummy '{test_zip_name}' (Image based) for demonstration ---")
    dummy_test_src_dir = 'dummy_test_images_src'
    num_test_classes_dummy = 5 # Example classes in test.zip
    num_images_per_class_dummy = 5

    with zipfile.ZipFile(test_zip_name, 'w') as zipf:
        for class_id in range(num_test_classes_dummy):
            class_folder_name = f'test_class_{chr(ord("A") + class_id)}' # e.g., test_class_A
            os.makedirs(os.path.join(dummy_test_src_dir, class_folder_name), exist_ok=True)
            for i in range(num_images_per_class_dummy):
                img_name = f'img_{class_id}_{i}.png'
                img_path_in_src = os.path.join(dummy_test_src_dir, class_folder_name, img_name)
                # Create grayscale dummy image files for test.zip
                img_pil = Image.new('L', (IMG_WIDTH, IMG_HEIGHT), color=np.random.randint(0, 50)) # Dark background
                draw = ImageDraw.Draw(img_pil)
                if class_id % 3 == 0: draw.ellipse((5,5,23,23), fill=200) # Circle
                elif class_id % 3 == 1: draw.rectangle((5,5,23,23), fill=200) # Square
                else: draw.line((5,5,23,23), fill=200, width=2) # Line
                img_pil.save(img_path_in_src)

                zipf.write(img_path_in_src, os.path.join(class_folder_name, img_name)) # Write with folder structure in zip
    shutil.rmtree(dummy_test_src_dir)
    print(f"Dummy '{test_zip_name}' created with {num_test_classes_dummy} classes and {num_images_per_class_dummy} images each.")


# --- Load Data from train.csv and test.zip ---

# Load training data from train.csv
print(f"\n--- Loading training data from '{train_csv_name}' ---")
df_train = pd.read_csv(train_csv_name)
print(f"Training data shape (from CSV): {df_train.shape}")

# Display DataFrame head
print("\n--- Training Data (df_train) Head ---")
print(df_train.head())

# Determine number of classes and map labels (MOVED UP FOR SCOPE)
# This part assumes your 'label' column in train.csv contains numerical labels.
unique_train_labels = sorted(df_train['label'].unique().tolist())
NUM_CLASSES = len(unique_train_labels)

# Create a mapping from numerical label (0-24) to actual character (A-Z, skipping J/Z)
ASL_ALPHABET = [chr(ord('A') + i) for i in range(26)] # A, B, C, ... Z
CLASS_NAMES_MAP = unique_train_labels # Keep this for ticks/labels in plots
CLASS_NAMES = []
# Custom logic to map numerical labels to ASL characters or generic Class_X
if 9 not in unique_train_labels and NUM_CLASSES == 24: # Heuristic for standard ASL MNIST
    temp_asl = ASL_ALPHABET[:]
    temp_asl.pop(9) # Remove 'J'
    for label_val in unique_train_labels:
        # Assuming unique_train_labels are 0-8, 10-24
        # This re-maps 10 to 'K', 11 to 'L', etc.
        mapped_idx = label_val if label_val < 9 else label_val - 1
        CLASS_NAMES.append(temp_asl[mapped_idx])
    print(f"Detected {NUM_CLASSES} unique classes, mapped to ASL A-Z (excluding J): {CLASS_NAMES}")
elif len(unique_train_labels) == 26 and all(i in unique_train_labels for i in range(26)): # Full A-Z, no skips
    CLASS_NAMES = ASL_ALPHABET
    print(f"Detected {NUM_CLASSES} unique classes, mapped to ASL A-Z: {CLASS_NAMES}")
else: # Fallback for any other numerical label set
    CLASS_NAMES = [f'Class_{l}' for l in unique_train_labels]
    print(f"Using generic class names: {CLASS_NAMES}")

print(f"Identified {NUM_CLASSES} classes in training data: {CLASS_NAMES}")


# Preprocess training data
y_train_raw = df_train['label']
x_train_raw_pixels = df_train.drop('label', axis=1).values

# Normalize pixel values
x_train = x_train_raw_pixels / 255.0

# Reshape data to (num_samples, IMG_WIDTH, IMG_HEIGHT, CHANNELS)
x_train = x_train.reshape(-1, IMG_WIDTH, IMG_HEIGHT, CHANNELS)

# One-hot encode labels for training
label_binarizer = LabelBinarizer()
# Fit transform on all unique labels to handle cases where train/test labels might differ slightly
label_binarizer.fit(unique_train_labels) # Fit on all unique labels found
y_train_onehot = label_binarizer.transform(y_train_raw)

# Split training data into training and validation sets
x_train, x_val, y_train_onehot, y_val_onehot = train_test_split(
    x_train, y_train_onehot,
    test_size=VAL_SPLIT_FROM_TRAIN_CSV,
    random_state=42,
    stratify=y_train_raw # Stratify based on original labels to maintain class distribution
)

print(f"x_train shape: {x_train.shape}, y_train_onehot shape: {y_train_onehot.shape}")
print(f"x_val shape: {x_val.shape}, y_val_onehot shape: {y_val_onehot.shape}")


# Extract test data from test.zip using ImageDataGenerator
print(f"\n--- Extracting and setting up test data from '{test_zip_name}' ---")
# Use a temporary directory to extract contents of test.zip
os.makedirs(extracted_test_dir, exist_ok=True)
with zipfile.ZipFile(test_zip_name, 'r') as zip_ref:
    zip_ref.extractall(extracted_test_dir)
print(f"Extracted '{test_zip_name}' to '{extracted_test_dir}'.")

# Use ImageDataGenerator for test data
test_datagen = ImageDataGenerator(rescale=1./255) # Only rescale, no augmentation for test

test_generator = test_datagen.flow_from_directory(
    extracted_test_dir,
    target_size=(IMG_WIDTH, IMG_HEIGHT),
    color_mode='grayscale' if CHANNELS == 1 else 'rgb',
    batch_size=BATCH_SIZE,
    class_mode='categorical', # Expected by model for evaluation
    shuffle=False # Crucial for consistent evaluation and matching filenames/labels
)

print(f"Found {test_generator.samples} test images belonging to {test_generator.num_classes} classes from '{test_zip_name}'.")

# Store the true labels from the test generator for later use
# test_generator.classes gives the numerical class indices for each image in order
y_test_true_indices = test_generator.classes


# --- 2. Data Visualization ---

# 2.1. Initial Sample Image Grid from Training Data
print("\n--- Initial Data Visualization (Sample Images from Training Set) ---")
plt.figure(figsize=(10, 5)) # Adjusted figure size
num_display_samples_initial = 10 # Display 10 samples (2 rows, 5 columns)
num_display_samples_initial = min(num_display_samples_initial, x_train.shape[0]) # Don't exceed available samples

# Randomly select indices for display
display_indices_initial = np.random.choice(x_train.shape[0], size=num_display_samples_initial, replace=False)

for i, idx in enumerate(display_indices_initial):
    ax = plt.subplot(2, 5, i + 1)
    img_display = x_train[idx].reshape(IMG_WIDTH, IMG_HEIGHT) # Already normalized
    plt.imshow(img_display, cmap='gray', vmin=0, vmax=1) # Normalized pixels 0-1
    plt.title(f"Label: {CLASS_NAMES[np.argmax(y_train_onehot[idx])]}", fontsize=9)
    plt.axis('off')
plt.suptitle('Sample Images from Training Data (Normalized)', fontsize=14)
plt.tight_layout(rect=[0, 0.03, 1, 0.95])
plt.show()
print(f"Displayed {num_display_samples_initial} sample images from the normalized training data.")


# 2.2. Class Distribution Bar Chart
print("\n--- Class Distribution in Training Data ---")
plt.figure(figsize=(10, 6))
# Use `y_train_raw` for countplot as it contains original numerical labels
sns.countplot(x=y_train_raw, palette='viridis')
plt.title('Distribution of Classes in Training Data')
plt.xlabel('Class Label')
plt.ylabel('Count')
plt.xticks(ticks=CLASS_NAMES_MAP, labels=CLASS_NAMES) # Map numerical labels to actual names
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.show()


# --- 3. Model Architecture ---
print("\n--- Defining Model Architecture ---")
model = Sequential()
model.add(Conv2D(75 , (3,3) , strides = 1 , padding = 'same' , activation = 'relu' , input_shape = (IMG_WIDTH,IMG_HEIGHT,CHANNELS)))
model.add(BatchNormalization())
model.add(MaxPooling2D((2,2) , strides = 2 , padding = 'same')) # Changed MaxPool2D to MaxPooling2D
model.add(Conv2D(50 , (3,3) , strides = 1 , padding = 'same' , activation = 'relu'))
model.add(Dropout(0.2))
model.add(BatchNormalization())
model.add(MaxPooling2D((2,2) , strides = 2 , padding = 'same')) # Changed MaxPool2D to MaxPooling2D
model.add(Conv2D(25 , (3,3) , strides = 1 , padding = 'same' , activation = 'relu'))
model.add(BatchNormalization())
model.add(MaxPooling2D((2,2) , strides = 2 , padding = 'same')) # Changed MaxPool2D to MaxPooling2D
model.add(Flatten())
model.add(Dense(units = 512 , activation = 'relu'))
model.add(Dropout(0.3))
model.add(Dense(units = NUM_CLASSES , activation = 'softmax')) # Use NUM_CLASSES here
model.compile(optimizer = 'adam' , loss = 'categorical_crossentropy' , metrics = ['accuracy'])
model.summary()


# --- 4. Model Training ---
print(f"\n--- Starting Model Training for {EPOCHS} epochs ---")

# Data augmentation for training data (x_train, y_train_onehot)
datagen = ImageDataGenerator(
    featurewise_center=False,
    samplewise_center=False,
    featurewise_std_normalization=False,
    samplewise_std_normalization=False,
    zca_whitening=False,
    rotation_range=10,
    zoom_range = 0.1,
    width_shift_range=0.1,
    height_shift_range=0.1,
    horizontal_flip=False,
    vertical_flip=False
)
datagen.fit(x_train)

# Learning rate reduction callback
learning_rate_reduction = ReduceLROnPlateau(monitor='val_accuracy', patience=2, verbose=1, factor=0.5, min_lr=0.00001)

history = model.fit(
    datagen.flow(x_train, y_train_onehot, batch_size=BATCH_SIZE),
    epochs=EPOCHS,
    validation_data=(x_val, y_val_onehot), # Validate using x_val, y_val_onehot
    callbacks=[learning_rate_reduction],
    verbose=0 # Suppress verbose output during training
)

print("\nModel training finished.")

# --- 5. Evaluation and Plots ---

# 5.1. Evaluation on Validation Set and Plot History
print("\n--- Evaluating on Validation Set ---")
val_loss, val_accuracy = model.evaluate(x_val, y_val_onehot, verbose=0)
print(f"Validation Loss: {val_loss:.4f}")
print(f"Validation Accuracy: {val_accuracy:.4f}")

epochs_range = [i for i in range(EPOCHS)] # Use EPOCHS from config
fig , ax = plt.subplots(1,2)
train_acc = history.history['accuracy']
train_loss = history.history['loss']
val_acc = history.history['val_accuracy']
val_loss = history.history['val_loss']
fig.set_size_inches(16,9)

ax[0].plot(epochs_range , train_acc , 'go-' , label = 'Training Accuracy')
ax[0].plot(epochs_range , val_acc , 'ro-' , label = 'Validation Accuracy')
ax[0].set_title('Training & Validation Accuracy')
ax[0].legend()
ax[0].set_xlabel("Epochs")
ax[0].set_ylabel("Accuracy")

ax[1].plot(epochs_range , train_loss , 'g-o' , label = 'Training Loss')
ax[1].plot(epochs_range , val_loss , 'r-o' , label = 'Validation Loss')
ax[1].set_title('Training & Validation Loss') # Corrected title
ax[1].legend()
ax[1].set_xlabel("Epochs")
ax[1].set_ylabel("Loss")
plt.show()
print("Validation evaluation and history plots complete.")

# 5.2. Test Run on Unseen Test Set (from test.zip images)
print("\n--- Performing Test Run on Unseen Test Set (from test.zip images) ---")
test_loss, test_accuracy = model.evaluate(test_generator, verbose=0)
print(f"Test Loss: {test_loss:.4f}")
print(f"Test Accuracy: {test_accuracy:.4f}")
print(f"Accuracy of the model is - {test_accuracy*100:.2f}%") # Kaggle style output

# Get predictions for detailed metrics
test_generator.reset() # Reset generator to ensure predictions align with labels/filenames
predictions_prob_test = model.predict(test_generator, verbose=0)
predicted_classes_test = np.argmax(predictions_prob_test, axis=-1)

# Handle the potential 'J' skip for ASL if your labels are 0-24, skipping 9
# This assumes the test_generator's classes align with the train_csv labels (0-24 without J)
# If your test.zip has different labels or doesn't follow the ASL MNIST convention, adjust this.
if 'J' not in CLASS_NAMES and NUM_CLASSES == 24: # This heuristic implies ASL MNIST like labels
    # If the class labels were 0-24 but 9 was skipped (like ASL MNIST),
    # predictions might need to be adjusted to align with a full 0-25 alphabet for reporting.
    # The original Kaggle notebook had this logic.
    # However, if CLASS_NAMES is derived directly from unique_train_labels (0-24, 1-10, etc),
    # then no adjustment to predicted_classes_test is needed, just ensure target_names is correct.
    # Let's rely on CLASS_NAMES for target_names.
    pass # No change needed if CLASS_NAMES correctly reflects unique labels
    # For a direct mapping from 0-24 (without J) to ASL_ALPHABET characters (A-Y without J),
    # you'd need `ASL_CLASSES.pop(9)` logic on a 0-25 set, then use `ASL_CLASSES` here.
    # The current CLASS_NAMES is based on actual unique labels, which is more robust.


print("\n--- Classification Report on Test Set ---")
# Use CLASS_NAMES (derived from train.csv) for target_names, which handles the correct label names
print(classification_report(y_test_true_indices, predicted_classes_test, target_names=CLASS_NAMES))


print("\n--- Confusion Matrix on Test Set ---")
cm_test = confusion_matrix(y_test_true_indices, predicted_classes_test)

# Convert to DataFrame for Seaborn heatmap as in Kaggle script
# Use CLASS_NAMES for both index and columns for clarity
cm_df = pd.DataFrame(cm_test , index = CLASS_NAMES , columns = CLASS_NAMES)

plt.figure(figsize = (15,15))
sns.heatmap(cm_df, cmap= "Blues", linecolor = 'black' , linewidth = 1 , annot = True, fmt='')
plt.xlabel('Predicted Label')
plt.ylabel('True Label')
plt.title('Confusion Matrix for Test Set')
plt.show()
print("Test run metrics and confusion matrix generated.")


# --- 7. Display Correctly Predicted Images from Test Set ---
print("\n--- Displaying Correctly Predicted Sample Images from Test Set ---")

num_correct_samples_to_display = 6 # Display 6 samples as in your reference image
if test_generator.samples == 0:
    print("No test samples available to display from test.zip.")
else:
    # Find indices where prediction matches true label
    correctly_predicted_indices = np.where(y_test_true_indices == predicted_classes_test)[0]

    if len(correctly_predicted_indices) == 0:
        print("No correctly predicted samples found in the test set to display.")
    else:
        # Randomly select a subset of correctly predicted indices for display
        display_correct_indices = np.random.choice(
            correctly_predicted_indices,
            size=min(num_correct_samples_to_display, len(correctly_predicted_indices)),
            replace=False
        )

        plt.figure(figsize=(10, 9)) # Adjusted figure size for 3x2 grid

        for i, idx in enumerate(display_correct_indices):
            plt.subplot(3, 2, i + 1) # 3 rows, 2 columns

            # Get the full path to the original image from the test set
            image_relative_path = test_generator.filenames[idx]
            full_image_path = os.path.join(extracted_test_dir, image_relative_path)

            # Get true and predicted labels (numerical indices)
            true_label_idx = y_test_true_indices[idx]
            predicted_label_idx = predicted_classes_test[idx]

            # Display the image (load original for display)
            try:
                if os.path.exists(full_image_path):
                    original_img_display = Image.open(full_image_path)

                    # Convert to grayscale if model expects 1 channel
                    if CHANNELS == 1 and original_img_display.mode != 'L':
                        original_img_display = original_img_display.convert('L')
                    elif CHANNELS == 3 and original_img_display.mode != 'RGB':
                        original_img_display = original_img_display.convert('RGB')

                    original_img_display = original_img_display.resize((IMG_WIDTH, IMG_HEIGHT))

                    # Display the image with appropriate colormap
                    if original_img_display.mode == 'L':
                        plt.imshow(original_img_display, cmap='gray', vmin=0, vmax=255)
                    else:
                        plt.imshow(original_img_display)
                else:
                    print(f"Warning: Original image file not found at {full_image_path}. Displaying placeholder.")
                    plt.text(IMG_WIDTH/2, IMG_HEIGHT/2, "Image Not Found", horizontalalignment='center', verticalalignment='center', color='red', fontsize=10)
                    plt.imshow(np.zeros((IMG_WIDTH, IMG_HEIGHT, CHANNELS), dtype=np.uint8).squeeze()) # Black square fallback
            except Exception as e:
                print(f"Error loading/displaying image from path {full_image_path}: {e}")
                plt.text(IMG_WIDTH/2, IMG_HEIGHT/2, "Error Loading Image", horizontalalignment='center', verticalalignment='center', color='red', fontsize=10)
                plt.imshow(np.zeros((IMG_WIDTH, IMG_HEIGHT, CHANNELS), dtype=np.uint8).squeeze()) # Black square fallback

            # Title format: "Predicted Class X, Actual Class Y"
            # Use CLASS_NAMES to get descriptive labels
            title_text = f"Predicted Class {CLASS_NAMES[predicted_label_idx]}, Actual Class {CLASS_NAMES[true_label_idx]}"
            plt.title(title_text, fontsize=10)
            plt.axis('off')

        plt.tight_layout()
        plt.show()
        print("Sample correctly predicted test images displayed.")

print("\n--- Full Hand Gesture Recognition Process Complete ---")
print(f"Model saved at: {model_save_path}")

# --- Final Cleanup ---
print("\n--- Final Cleanup ---")
if os.path.exists(extracted_train_dir_temp):
    shutil.rmtree(extracted_train_dir_temp)
    print(f"Cleaned up '{extracted_train_dir_temp}'.")
if os.path.exists(extracted_test_dir):
    shutil.rmtree(extracted_test_dir)
    print(f"Cleaned up '{extracted_test_dir}'.")
print("Cleanup complete.")
```
